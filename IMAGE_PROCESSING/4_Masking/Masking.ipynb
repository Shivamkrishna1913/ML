# Thresholding on Grayscale

```python
 import cv2 as cv

cam = cv.VideoCapture(0)

logo = cv.imread('Downloads/logo.jpg') 

while True:

    _,img = cam.read()
    img = cv.flip(img,1)  
    img = cv.resize(img,(300,200)) 
    
    gray = cv.cvtColor(img , cv.COLOR_BGR2GRAY) ## convert to gray
    img[img<200] = 0
    threshold_1 = gray.copy()
    threshold_1[threshold_1>200]=40   ## pixel shines more will become almost black

    threshold_2 = gray.copy()
    threshold_2[threshold_2>200]=150   ## pixel shines more will become less white

    
    
    # img[0:100,540:640] = logo # top-right
    # img[0:100,0:100] = logo # top-left
    # img[380:480,540:640] = logo # bottom-right
    # img[380:480,0:100] = logo # bottom-left
    ht = img.shape[0]
    wt = img.shape[1]

    # top-left
    margin_top  = int(ht/100)             # 1%
    logo_height = int((ht/100)*11)        # 11%
    margin_left = int((wt/100))           # 1%
    logo_width  = int((wt/100)*8)         # 8%

    # bottom-left
    margin_top  = int((ht/100)*88)        # 88%
    logo_height = int((ht/100)*11)        # 11%
    margin_left = int((wt/100))           # 1%
    logo_width  = int((wt/100)*8)         # 8%

    # bottom-right
    margin_top  = int((ht/100)*88)        # 88%
    logo_height = int((ht/100)*11)        # 11%
    margin_left = int((wt/100)*91)        # 91%
    logo_width  = int((wt/100)*8)         # 8%

    # top-right
    margin_top  = int((ht/100)*1)         # 1%
    logo_height = int((ht/100)*11)        # 11%
    margin_left = int((wt/100)*91)        # 91%
    logo_width  = int((wt/100)*8)         # 8%

    bg = img[margin_top : logo_width + margin_top , margin_left : margin_left + logo_width].copy()

    logo = cv.resize(logo, (logo_width, logo_width))

    final_logo = cv.addWeighted(bg,0.3,logo,0.7,0)
    if margin_left + logo_width <= wt and logo_width + margin_top<= ht :
       img[ margin_top : logo_width + margin_top , margin_left : margin_left + logo_width ] = final_logo
    else :
        print("logo doesn't fit in frame -- ")

    cv.imshow('Frame',img)
    cv.imshow('Gray',gray)
    cv.imshow('Threshold_1',threshold_1)
    cv.imshow('threshold_2',threshold_2)
    
    if cv.waitKey(1)==27:
        cam.release()
        cv.destroyAllWindows() 
        break 
```
