```python
import re
import numpy as np
import pandas as pd
```

```python
with open('big.txt','r') as fd:
  lines = fd.readlines()
  words=[]
  for line in lines:
    words+=re.findall('\w+',line.lower())
```

Making the pairs

```python
def get_pairs(words,n):

  data=[]
  for i in range(len(words)-n):
     data.append(' '.join(words[i:i+n]))
  return data

def prob_dist(data):
  a=np.array(data)
  pair,count=np.unique(a,return_counts=True)

  prob_dist=[]
  for i in range(len(pair)):
    b = pair[i].split(' ')
    left_seq = ' '.join(b[:len(b)//2])
    right_seq = ' '.join(b[len(b)//2+1:])
    middle =  b[len(b)//2]

    prob_dist.append([pair[i],left_seq,right_seq,middle,count[i]])
  return prob_dist

data = get_pairs(words,5)
p_dist = prob_dist(data)

df=pd.DataFrame(p_dist,columns=['seq','in','out','mid','freq'])
df.head()
```

```python
word = 'in the _ of the'
left_seq = word.split('_')[0].strip()
right_seq = word.split('_')[1].strip()

df_ = df[df['in']==left_seq]
df_ = df_[df_['out']==right_seq]
list(df_.sort_values(by='freq',ascending=False).head()['mid'].values)
```
