```python
text = "Steve Jobs was a visionary entrepreneur and co-founder of Apple Inc., whose influence reshaped the technology industry. Born in 1955, Jobs was known for his relentless drive for innovation and his ability to anticipate consumer needs before they became apparent. Under his leadership, Apple introduced groundbreaking products like the iPhone, iPad, and MacBook, which revolutionized the way people interact with technology. Jobs' emphasis on design, simplicity, and user experience set new industry standards. Despite facing challenges, including being ousted from Apple in the 1980s, Jobs returned to lead the company to unprecedented success. His legacy endures as a symbol of creativity and relentless pursuit of excellence."
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopword = stopwords.words('english')
```

```python
wc = WordCloud(width=1000,height=750,margin=2,max_words=100,background_color='white',stopwords=stopword).generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
