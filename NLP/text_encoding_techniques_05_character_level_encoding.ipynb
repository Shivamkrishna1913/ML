With CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
from pandas import DataFrame
text_data=['GFG is providing a new learn Deep learning course which is really good.',
           'We will be studying deep learning from today.',
           'I want a deep sleep today.','ziz big']
vec = CountVectorizer(analyzer='char' )
data_transformed = vec.fit_transform(text_data)

# print(vec.vocabulary_)
# print(data_transformed.toarray())
df=DataFrame(data_transformed.toarray(),columns=vec.vocabulary_)
df
```

encoding from scratch

```python
text_data=['GFG is providing a new learn Deep learning course which is really good.',
           'We will be studying deep learning from today.',
           'I want a deep sleep today.','ziz big']
characters = list(set(''.join(text_data).lower()))
char_to_int = { char:idx for idx,char in enumerate(characters)}

enc_data = [[char_to_int[char] for char in sent.lower()] for sent in text_data]
print(enc_data)
```

decoding

```python
int_to_char = { idx:char for idx,char in enumerate(characters)}

dec_data = [''.join([int_to_char[char] for char in sent]) for sent in enc_data]
print(dec_data)
```
